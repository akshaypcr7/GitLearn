# EDA


### uni variable analysis

```python
df['rating']
```

```python
datae = df[df['price'] < 1000]
plt.title('Price Distribution')
sns.histplot(data=datae, x='price', bins=50)
plt.ylabel('Frequency')
plt.figure(figsize=(6,5))
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.histplot(data=datae, x='reviews_per_month', bins=100)
plt.ylabel('Frequency')
plt.title('Reviews')
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.histplot(data=datae, x='availability_365')
plt.ylabel('Frequency')
plt.title('Avail')
plt.show()
```

```python
plt.title('Price compare for room type')
sns.barplot(data=datae,x='neighbourhood_group',y='price',hue='room_type')
plt.figure(figsize=(6,8))
plt.show()
```

```python
plt.title('Price compare based on reviews')
sns.scatterplot(data=datae,x='number_of_reviews',y='price',hue='neighbourhood_group')
plt.show()
```
